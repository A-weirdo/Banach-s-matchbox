```python
import random
import numpy as np
from matplotlib import pyplot as plt
import math
I=[]
N=20

times=100000
for n in range(0 , times):
 N1=N
 N2=N

 while 1 :
     if random.randint(1, 2) == 1 :
       N1=N1-1
     else :
       N2=N2-1
     if N1 == 0 :
       k=N2
       I.append(k)
       break
     if N2 == 0 :
       k=N1
       I.append(k)
       break
  
sum=0
E=0
for n in range(0 , N+1):
   E+=n*I.count(n)/ times
   
print("E_theory = " + str(2*N*math.factorial(N*2)/(math.factorial(N)**2)*(0.5)**(2*n)))
print("E_simulate = " + str(E))


print ("for " +str(times) + " simulation we have :")
for n in range (1, N+1):
   print ("if k= " + str(n)+   "  the probability of simulation is  "    +str(I.count(n)/ times))
   m=math.factorial(N*2-n-1)/(math.factorial(N-1)*math.factorial(N-n))*(0.5)**(N*2-n-1)
   print ("if k= " + str(n)+   "  the probability of theory is  "    +str(m))
   sum += m




plt.hist(I, len(set(I)))
plt.xlabel("times = "+ str(times))
plt.ylabel("y-axis")
plt.title('N = ' + str(N))
plt.show()


```
